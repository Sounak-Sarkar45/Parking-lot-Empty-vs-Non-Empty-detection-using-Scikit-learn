```python
import os
import pickle
from skimage.io import imread
from skimage.transform import resize
import numpy as np
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
input_dir='clf-data'
categories=['empty','not_empty']
data=[]
labels=[]
for category_idx,category in enumerate(categories):
    for file in os.listdir(os.path.join(input_dir,category)):
        img_path=os.path.join(input_dir,category,file)
        img=imread(img_path)
        img=resize(img,(15,15))
        data.append(img.flatten())
        labels.append(category_idx)
```

```python
data=np.asarray(data)
labels=np.asarray(labels)
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,labels,test_size=0.25,shuffle=True,stratify=labels,random_state=44)
```

```python
classifier=SVC()
```

```python
parameters=[{'gamma':[0.01,0.001,0.0001],'C':[1,10,100,1000]}]
grid_search=GridSearchCV(classifier,parameters)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_estimator=grid_search.best_estimator_
```

```python
y_pred=best_estimator.predict(x_test)
```

```python
print('{}% of samples are correctly classified'.format(str(accuracy_score(y_pred,y_test)*100)))
```

```python
pickle.dump(best_estimator,open('./model.p','wb'))
```
